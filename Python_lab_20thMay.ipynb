```python
data['Date Time']=pd.to_datetime(data['Date']+' '+data['Time'])
daily_sales=data.groupby(data['Date Time'].dt.date)['Total'].sum()
plt.figure(figsize=(10,5))
plt.plot(daily_sales.index,daily_sales.values, marker='*')
plt.title('Daily Sales Trends')
plt.xlabel('Data')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
payment_methods=data['Payment'].value_counts()

plt.figure(figsize=(10,8))
payment_methods.plot(kind='bar')
plt.xlabel('Payment Method')
plt.ylabel('CNumber of Transactions')
plt.title('Payment Methods Distribution')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
gender_counts=data['Gender'].value_counts()
plt.figure(figsize=(10,8))
gender_counts.plot(kind='pie',autopct='%1.1f%%',startangle=90)
plt.xlabel('Gender')
plt.ylabel('Number of Transactions')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.title('Gender Distribution')
plt.show()
```

```python
customer_type_counts=data['Customer type'].value_counts()

plt.figure(figsize=(10,8))
customer_type_counts.plot(kind='bar')
plt.xlabel('Customer Type')
plt.ylabel('Number of Customers')
plt.title('Customer Type Distribution')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```
